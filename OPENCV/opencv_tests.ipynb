```python
import cv2
import matplotlib.pyplot as plt
```

```python
# imread just read the image from the file directory

img = cv2.imread('pic1.jpg')

print(img.shape)
```

```python
# for show the image
# it returns the pic in BGR format but we have to convert it in a RGB format

plt.imshow(img)
plt.show()
```

```python
#converting BGR to RGB

img = cv2.imread('pic1.jpg')
rgb_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_image)
plt.show()
```

```python
# for turn off the axis

plt.imshow(rgb_image)
plt.axis("off")
plt.show()
```

```python
# Crop image [row,column,all_channels]

crop_image = rgb_image[0:1000,200:1200,:]
plt.imshow(crop_image)
plt.axis("off")
plt.show()
```

```python
crop_image.shape
```

```python
total_pixels = crop_image.shape[0]*crop_image.shape[1]

print(total_pixels)
```

```python
# resize the image

resize_image = cv2.resize(crop_image,(100,100))

plt.imshow(resize_image)
plt.axis("off")
plt.title("kaju"+str(resize_image.shape))
plt.show()
```

```python
total_pixels = resize_image.shape[0]*resize_image.shape[1]

print(total_pixels)
```
