```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# for searching the ploting style

plt.style.available
```

```python
x = np.array([1,2,3,4,5])
y = x**2 + 3
```

```python
print(x,y)
```

```python
plt.style.use("dark_background")
plt.plot(x,y)     # for line type plotting "plot" used
plt.title("My line plot")    # give a title to the graph
plt.show()
```

```python
# give the label to the graph
# xlabel and ylebel method is for labelling the axix

plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.plot(x,y+x,label = "Stock Price 2")
plt.plot(x,y,label = "Stock Price 1")
plt.title("My line plot")    # give a title to the graph
plt.legend()    # for give the lebel
plt.show()
```

```python
# For dot type plotting "scatter" plot used 

plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.scatter(x,y+x,label = "Stock Price 2")
plt.scatter(x,y,label = "Stock Price 1")
plt.title("My doted plot")    # give a title to the graph
plt.legend()    # for give the lebel
plt.show()
```
